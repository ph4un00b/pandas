```python
# Import necessary libraries
import pandas as pd

# Load data files into DataFrames
week1_sales = pd.read_csv('Restaurant - Week 1 Sales.csv')
week1_satisfaction = pd.read_csv('Restaurant - Week 1 Satisfaction.csv')
week2_sales = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

## outer join

```python
merged_sales = week1_sales.merge(
    week2_sales, 
    on=['Customer ID', 'Food ID'], 
    how='outer', 
    suffixes=('_week1', '_week2'),
    indicator=True
)
print("\na: ")
print(week1_sales.head())
print("\nb: ")
print(week2_sales.head())
merged_sales
```

## filter by indicator

```python
both = merged_sales[merged_sales['_merge'] == 'both']
both
```
