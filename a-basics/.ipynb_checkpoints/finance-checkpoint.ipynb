```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
yf.__version__
```

```python
ticker = 'MSFT'
```

```python
data = yf.Ticker(ticker)
h = data.history(period='id', start='2018-01-01', end='2024-01-01')
print(h.head())
h.info()
```

```python
closing = h['Close']
plt.plot(closing)
plt.title('moco clocing prices')
plt.xlabel('date')
plt.ylabel('usd')

plt.legend(['jamon legend'])
plt.show()
```

```python
closing = h['Close']
# plt.plot(closing)
# plt.title('moco closing prices
# Create a new figure and set the size
plt.figure(figsize=(10, 6))
# Plot the closing prices with a blue solid line, markers, and a custom line width
plt.plot(closing, color='blue', marker='.', linestyle='-', linewidth=.2, label='mocosoft')
plt.title('mocosft')
plt.xlabel('date')
plt.ylabel('usd')

plt.legend(['jamon legend'])
plt.grid(True)
# Customize x-axis ticks to display one tick per year
years = np.arange(closing.index.min().year, 2024 + 1)
tick_positions = [f'{year}-01-01' for year in years]
tick_labels = [str(year) for year in years]
plt.xticks(tick_positions, tick_labels)
# Customize y-axis ticks with a step of 50
y_tick_positions = np.arange(0, closing.max() + 100, 50)
plt.yticks(y_tick_positions)

plt.show()
```

```python
# Create a new figure and set the size
plt.figure(figsize=(19, 6))

# Set the background color to purple
plt.gca().set_facecolor('purple')

# Plot the closing prices with a purple solid line, smaller markers, and a custom line width
plt.plot(closing, color='whitesmoke', marker='o', markersize=3, linestyle='-', linewidth=1.0, label='mocosoft')

# Set the title and labels for the plot
plt.title('Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')

# Customize x-axis ticks to display one tick per month
month_range = pd.date_range(closing.index.min(), closing.index.max(), freq='ME')
tick_positions = month_range.strftime('%Y-%m-%d').tolist()
tick_labels = month_range.strftime('%Y-%m').tolist()
plt.xticks(tick_positions, tick_labels, rotation=90)

# Customize y-axis ticks with a step of 100
y_tick_positions = np.arange(0, closing.max() + 100, 100)
plt.yticks(y_tick_positions)

# Add a legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
