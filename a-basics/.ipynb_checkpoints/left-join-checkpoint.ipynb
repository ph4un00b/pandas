```python
# Import necessary libraries
import pandas as pd

# Load data files into DataFrames
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week1_satisfaction = pd.read_csv('Restaurant - Week 1 Satisfaction.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

## left join❗

```python
info = week1.merge(foods, left_on='Food ID', right_on='Food ID', how='left')
info
```

```python
all = info.merge(customers, left_on='Customer ID', right_on='ID', how='left')
all
```

## a.join(b)

```python
w1 = week1.set_index(['Customer ID', 'Food ID'])
w2 = week2.set_index(['Customer ID', 'Food ID'])

combo = w1.join(w2, lsuffix='_week1', rsuffix='_week2')
combo
```
